```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

sp= pd.read_csv('studperform.csv')
sp.head() 
```

```python
if 'Extracurricular Activities' in sp.columns:
    sp['Extracurricular Activities']= sp['Extracurricular Activities'].map({'Yes': 1,  'No': 0})
    
sp.head()
```

```python
corrmatrix= sp.corr()
print('The correlation matrix is:', corrmatrix)
```

```python
#Visualization

plt.figure(figsize=(10,8))
plt.title('Correlation Heatmap')

sns.heatmap(data= corrmatrix, annot= True, cmap= 'plasma', fmt= '.2f')
```
